```python
fig,axes=plt.subplots(nrows=1,ncols=2)
ax1=fig.add_axes([0, 0, 1, 1])
ax2=fig.add_axes([0.2, 0.5,0.4,0.4])
ax1.plot(x,z)
ax2.plot(x,y)
for ax in axes:
    ax1.set_xlabel("x")
    ax1.set_ylabel("z")
    ax1.set_title("x vs x**2")
    ax2.set_xlabel("x")
    ax2.set_ylabel("y")
    ax2.set_title("x vs x*2")
fig
```
